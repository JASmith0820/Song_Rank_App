# We scraped data from the following websites:

### 1) The Billboard Hot 100:
### Ranks songs using a combination of radio airplay impressions and sales data from Nielson Music, as well as metrics derived from online streaming data.

http://www.billboard.com/charts/hot-100

### 2) The Billboard Trending 140 in the Past 24 Hours:
### Real-time rankings of the most-shared songs on Twitter in the past 24 hours.
http://realtime.billboard.com/?chart=last24hours

### 3) The Shazam Top 100:
### The most popular songs among users of the Shazam mobile application.
http://www.shazam.com/charts/top-100/united-states

### 4) The Shazam Hit Predictor:
### The top 20 songs that executives at Shazam expect to be hits in the future.
http://www.shazam.com/charts/future-hits/united-states

### Note: All four of these data sources focus exlusively on music currently popular in North America.

# Billboard Hot 100 vs. Shazam Top 100
### Purpose: Determine if a relationship exists between the Billboard Hot 100 rankings and the Shazam Top 100 rankings
### Conclusion: Shazam rankings do not appear to correlate closely with the Billboard rankings. This is likely to be because Billboard incorporates a variety of factors into their rankings, including sales data and radio airplay impressions. Since Shazams ranking process appears not to take such a multi-faceted approach, it is likely that they cannot gauge music popularity as accurately as Billboard can.

```python
from IPython.display import Image
Image('BillboardHot100_vs_ShazamTop100.png')
```

# Shazam Hit Predictor vs. Billboard Hot 100
### Purpose: Determine if the Shazam Hit Predictor is accurate in predicting songs on the Billboard Hot 100 one week later
### Conclusion: Only 30% (6 out of 20) of the Shazam predicted hits songs made it onto the Billboard Hot 100, indicating that the Shazam hit predictor is not a particularly strong predictive tool, at least in the near term. 

```python
from IPython.display import Image
Image('Hit_Predictor_vs_Top_100.png')
```

# Billboard Top Trending vs. Billboard Hot 100
### Purpose: Determine if the Billboard Top Trending is accurate in predicting songs on the Billboard Hot 100 one week later
### Note: Although 140 songs are available on the Billboard Top Trending list, we limited our data to only the top 20 songs so that it would be comparable with the Shazam Hit Predictor analysis.
### Conclusion: 75% (15 out of 20) of the Top Trending songs made it onto the Billboard Hot 100, indicating that the trending data that Billboard mines from Twitter is a strong predictor of emerging hits. 

```python
from IPython.display import Image
Image('Top_Trending_vs_Top_100.png')
```

# Artist Hometown Map

# Conclusions

### 1) The Shazam Top 100 is not highly correlated with the Billboard Hot 100 (r~.327).
### 2) The Shazam Hit Predictor is not currently a strong predictor of future hits on the Billboard Hot 100.
### 3) The Billboard Trending 140 is a strong predictor of future hits on the Billboard Hot 100, suggesting that the Twitter data mined by Billboard to create this chart may be very valuable in identifying emerging hits.
